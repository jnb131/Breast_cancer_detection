```python
# split dataframe in feature and target
X = df.iloc[:,2:31].values
Y = df.iloc[:,1].values
```

```python
# split the dataset into 75% training and 25% Testing

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
def models(X_train,Y_train):
    #Logistic Regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression()
    log.fit(X_train,Y_train)
    
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion ='entropy')
    tree.fit(X_train,Y_train)
    
    #Random forest
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators = 10,criterion ='entropy')
    forest.fit(X_train,Y_train)
    
    #print accuracy in training data
    print('[0]Logostic Regression training accuracy: ',log.score(X_train,Y_train))
    print('[0]Decision Tree training accuracy: ',tree.score(X_train,Y_train))
    print('[0]Random forest training accuracy: ',forest.score(X_train,Y_train))
    
    return log,tree,forest
    
```

```python
model=models(X_train,Y_train)
```

```python
#test model accuracy on test data on each model
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print('model ',i)
    cm = confusion_matrix(Y_test,model[i].predict(X_test))
    TP = cm[0][0]
    FP = cm[0][1]
    FN = cm[1][0]
    TN = cm[1][1]
    print('Testing Accuracy:',(TP+TN)/(TP+TN+FP+FN))
    print(cm)
```

```python
from sklearn.metrics import classification_report 
from sklearn.metrics import accuracy_score
for i in range(len(model)):
    print('model ',i)
    print()
    print( classification_report(Y_test,model[i].predict(X_test)))
```
